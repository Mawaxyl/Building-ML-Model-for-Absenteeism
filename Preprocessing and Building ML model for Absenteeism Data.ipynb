### Standardizing only the Numerical Variables (Creating a Custom Scaler)

```python
#from sklearn.preprocessing import StandardScaler

#absenteeism_scaler = StandardScaler()
```

```python
from sklearn.base import BaseEstimator, TransformerMixin 
from sklearn.preprocessing import StandardScaler

class CustomScaler(BaseEstimator, TransformerMixin):

    def __init__(self, columns):
        self.scaler = StandardScaler()
        self.columns = columns
        self.mean_ = None
        self.var_ = None

    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns], y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self

    def transform(self, X, y=None, copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self .columns]), columns=self.columns)
        X_not_scaled = X.loc[:, ~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled, X_scaled], axis=1)[init_col_order]
```

## Backward Elimination

The idea is that we can simplify our model by removing all features which have close to no contribution to the model

Usually, when we have the p-values, we get rid of all co-efficients with p-values > 0.05 (though we do not have p-value in sklearn)

If the weight is small enough, it won't make a difference anyway

So, if we remove these variables, the rest of our model should not really change in terms of coefficient values

Check the cell 70 and cell 108 above

Note; The elimination did not affect our model
